```python
%%writefile my_program.cpp


//  main.cpp
//  assignment8
//
//   Lizbeth Garcia
//

#include <iostream>
#include <vector>
using namespace std;

// B-Tree
template <class recType, int bTreeOrder>
struct bTreeNode {
    int recCount;
    vector<recType> keys;             // keys
    vector<bTreeNode*> children;      //  pointers
    bool isLeaf;

    bTreeNode(bool leaf = true) {
        recCount = 0;
        keys.resize(bTreeOrder - 1);
        children.resize(bTreeOrder, nullptr);
        isLeaf = leaf;
    }
};

template <class recType, int bTreeOrder>
class bTree {
private:
    bTreeNode<recType, bTreeOrder>* root;

    void splitChild(bTreeNode<recType, bTreeOrder>* parent, int index) {
        bTreeNode<recType, bTreeOrder>* fullChild = parent->children[index];
        bTreeNode<recType, bTreeOrder>* newChild = new bTreeNode<recType, bTreeOrder>(fullChild->isLeaf);
        int midIndex = (bTreeOrder - 1) / 2;

        for (int i = 0; i < bTreeOrder / 2; ++i)
            newChild->keys[i] = fullChild->keys[midIndex + 1 + i];

        if (!fullChild->isLeaf) {
            for (int i = 0; i <= bTreeOrder / 2; ++i)
                newChild->children[i] = fullChild->children[midIndex + 1 + i];
        }

        fullChild->recCount = midIndex;
        newChild->recCount = bTreeOrder / 2;

        for (int i = parent->recCount; i > index; --i) {
            parent->keys[i] = parent->keys[i - 1];
            parent->children[i + 1] = parent->children[i];
        }
        parent->keys[index] = fullChild->keys[midIndex];
        parent->children[index + 1] = newChild;
        parent->recCount++;
    }

    void insertNonFull(bTreeNode<recType, bTreeOrder>* node, recType key) {
        int i = node->recCount - 1;

        if (node->isLeaf) {
            while (i >= 0 && key < node->keys[i]) {
                node->keys[i + 1] = node->keys[i];
                i--;
            }
            node->keys[i + 1] = key;
            node->recCount++;
        } else {
            while (i >= 0 && key < node->keys[i])
                i--;

            i++;
            if (node->children[i]->recCount == bTreeOrder - 1) {
                splitChild(node, i);
                if (key > node->keys[i])
                    i++;
            }
            insertNonFull(node->children[i], key);
        }
    }

public:
    bTree() {
        root = new bTreeNode<recType, bTreeOrder>();
    }

    void insert(recType key) {
        if (root->recCount == bTreeOrder - 1) {
            bTreeNode<recType, bTreeOrder>* newRoot = new bTreeNode<recType, bTreeOrder>(false);
            newRoot->children[0] = root;
            splitChild(newRoot, 0);
            root = newRoot;
        }
        insertNonFull(root, key);
    }

    void inOrder() {
        inOrderTraversal(root);
        cout << endl;
    }

    void inOrderTraversal(bTreeNode<recType, bTreeOrder>* node) {
        if (node) {
            for (int i = 0; i < node->recCount; ++i) {
                inOrderTraversal(node->children[i]);
                cout << node->keys[i] << " ";
            }
            inOrderTraversal(node->children[node->recCount]);
        }
    }

    void printTree(bTreeNode<recType, bTreeOrder>* node = nullptr, int level = 0) {
        if (!node)
            node = root;

        cout << "Level " << level << ": ";
        for (int i = 0; i < node->recCount; ++i)
            cout << node->keys[i] << " ";
        cout << endl;

        if (!node->isLeaf) {
            for (int i = 0; i <= node->recCount; ++i)
                printTree(node->children[i], level + 1);
        }
    }
};

int main() {
    bTree<int, 3> tree;

    vector<int> keys = {10, 20, 5, 6, 12, 30, 7, 17};
    for (int key : keys)
        tree.insert(key);

    cout << "In-order traversal: ";
    tree.inOrder();

    cout << "Tree :" << endl;
    tree.printTree();

    return 0;
}

```
